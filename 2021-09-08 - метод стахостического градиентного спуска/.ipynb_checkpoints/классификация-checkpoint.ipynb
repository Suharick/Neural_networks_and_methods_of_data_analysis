```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

%matplotlib inline
```

```python
data = pd.read_csv('binary_classification.csv', parse_dates=True)
data.head()
```

```python
data1 = data[data['y'] == 1]
data0 = data[data['y'] == 0]
```

```python
fig, ax = plt.subplots()

ax.scatter(data1['X1'], data1['X2'], c = 'r')
ax.scatter(data0['X1'], data0['X2'], c = 'b')

fig.set_figwidth(8)   
fig.set_figheight(8)   

plt.show()
```

```python
array = data.values
X = array[:,0:2]
y = array[:,2]
```

```python
test_size = 0.2
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```
