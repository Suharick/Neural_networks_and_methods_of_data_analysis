## Задача 1

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rcParams['figure.figsize'] = (10,5)
```

```python
data = pd.read_csv('data/titanic.csv', index_col=0)
data.head()
```

```python
data['Sex'] = data['Sex'].replace({'male' : 0, 'female' : 1})
data['Embarked'] = data['Embarked'].replace({'S' : 0, 'C' : 1, 'Q' : 2})
data.head()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12,8))

sns.distplot(data['Survived'], 
             ax=axes[0, 0]
            )
axes[0, 0].set_title("Survived()", fontsize=16)


sns.distplot(data['Sex'], 
             ax=axes[1, 0]
            )
axes[1, 0].set_title("Sex()", fontsize=16)


sns.distplot(data['Age'], 
             ax=axes[0, 1]
            )
axes[0, 1].set_title("Age()", fontsize=16)


sns.distplot(data['SibSp'], 
             ax=axes[1, 1]
            )
axes[1, 1].set_title("SibSp()", fontsize=16)


sns.distplot(data['Pclass'], 
             ax=axes[2, 0]
            )
axes[2, 0].set_title("Pclass()", fontsize=16)


sns.distplot(data['Embarked'], 
             ax=axes[2, 1]
            )
axes[2, 1].set_title("Embarked()", fontsize=16)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.violinplot(
    x = 'Sex',
    y = 'Survived',
    data = data,
    ax=axes[0, 0]
);
axes[0, 0].set_title("Sex()", fontsize=16)


sns.violinplot(
    x = 'Embarked',
    y = 'Survived',
    data = data,
    ax=axes[0, 1]
);
axes[0, 1].set_title("Embarked()", fontsize=16)


sns.violinplot(
    x = 'SibSp',
    y = 'Survived',
    data = data,
    ax=axes[1, 0]
);
axes[1, 0].set_title("SibSp()", fontsize=16)


sns.violinplot(
    x = 'Pclass',
    y = 'Survived',
    data = data,
    ax=axes[1, 1]
);
axes[1, 1].set_title("Pclass()", fontsize=16)


plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.boxplot(
    x = 'Sex',
    y = 'Survived',
    data = data,
    ax=axes[0, 0]
);
axes[0, 0].set_title("Sex()", fontsize=16)


sns.boxplot(
    x = 'Embarked',
    y = 'Survived',
    data = data,
    ax=axes[0, 1]
);
axes[0, 1].set_title("Embarked()", fontsize=16)


sns.boxplot(
    x = 'SibSp',
    y = 'Survived',
    data = data,
    ax=axes[1, 0]
);
axes[1, 0].set_title("SibSp()", fontsize=16)


sns.boxplot(
    x = 'Pclass',
    y = 'Survived',
    data = data,
    ax=axes[1, 1]
);
axes[1, 1].set_title("Pclass()", fontsize=16)


plt.tight_layout()
```
