```python
import numpy as np
import pandas as pd

from PIL import Image
import glob
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
img = Image.open('train/indoor/3.jpg')
```

```python
np.asarray(img)[:,:,0]
```

```python
filenames1 = glob.glob("train\\indoor\\*")
```

```python
lst1 = []
for file in filenames1:
    try:
        img = Image.open(file)

        res = pd.DataFrame(np.asarray(img).reshape(3,-1).T).describe().values.reshape(-1).tolist()[3:-3]
        res.append(1)


        lst1.append(res)
    except:
        print(file+'error')
```

```python
lst1
```

```python
filenames0 = glob.glob("train\\outdoor\\*")
```

```python
lst0 = []
for file in filenames0:
    try:
        img = Image.open(file)

        res = pd.DataFrame(np.asarray(img).reshape(3,-1).T).describe().values.reshape(-1).tolist()[3:-3]
        res.append(0)


        lst0.append(res)
        
    except:
        print(file+'error')
```

```python
df0 = pd.DataFrame(lst0)
df1 = pd.DataFrame(lst1)
```

```python
df = df0.append(df1)
df
```

```python
array = df.values
X = array[:,0:18]
Y = array[:,18]
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=test_size, random_state=seed)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
result = model.score(X_test, Y_test)
```

```python
print('Accuracy: %.3f%%' %(result*100))
```
